```python
import os

# File to store data
LOG_FILE = "calorie_log.txt"

def load_data():
    meals = []
    if os.path.exists(LOG_FILE):
        with open(LOG_FILE, "r") as f:
            for line in f:
                parts = line.strip().split(",")
                if len(parts) == 2:
                    meals.append({"meal": parts[0], "calories": int(parts[1])})
    return meals

def save_data(meals):
    with open(LOG_FILE, "w") as f:
        for m in meals:
            f.write(f"{m['meal']},{m['calories']}\n")

def display_summary(meals, daily_limit):
    total = sum(m['calories'] for m in meals)
    remaining = daily_limit - total
    print(f"\n--- Daily Summary ---")
    print(f"Total Calories Consumed: {total}")
    print(f"Remaining Calories: {remaining if remaining > 0 else 0}")
    if total > daily_limit:
        print("  You've exceeded your daily limit!")
    print("-----------------------\n")

def main():
    print(" Welcome to Calorie Tracker CLI ")
    daily_limit = int(input("Enter your daily calorie limit: "))

    meals = load_data()

    while True:
        print("Choose an option:")
        print("1. Add Meal")
        print("2. View Summary")
        print("3. Exit")
        choice = input("Enter choice (1-3): ")

        if choice == "1":
            meal_name = input("Enter meal name: ")
            calories = int(input("Enter calories: "))
            meals.append({"meal": meal_name, "calories": calories})
            save_data(meals)
            print(" Meal added!\n")

        elif choice == "2":
            display_summary(meals, daily_limit)

        elif choice == "3":
            save_data(meals)
            print(" Data saved. Goodbye!")
            break

        else:
            print(" Invalid choice, try again.\n")

if __name__ == "__main__":
    main()
```
